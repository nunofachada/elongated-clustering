```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
sns.set_style("darkgrid")
```

```python
for i in range(1, 7):
    clusts = pd.read_csv(f"s{i}.csv")
    clusts['group'] = clusts['group'].astype(str)
    f = plt.figure()
    ax = sns.scatterplot(data=clusts, x="x", y="y", hue="group", legend=False)
    ax.set_title(f"S{i}: {len(clusts)} points, {len(clusts['group'].unique())} clusters")
    f.savefig(f"s{i}.svg")
```
