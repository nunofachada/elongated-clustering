```python
import pandas as pd
import numpy as np
```

```python
for i in range(1, 7):
    xy = pd.read_csv(f"S{i}.csv", names=('x','y'))
    grp = pd.read_csv(f"S{i}_clusters.csv", header=None)
    c = np.zeros(sum(grp[0]), dtype=np.int64)
    if sum(grp[0]) != len(xy):
        print("ERROR!")
    # else:
    #     print(len(xy))
    copied = 0
    for j in range(len(grp)):
        # print(f"\t{grp[0][j]}")
        npts = grp[0][j]
        c[copied:(copied + npts)] = j + 1
        copied += npts

    xy["group"] = c
    print(xy)
    xy.to_csv(f"s{i}.csv", index=None)
```
